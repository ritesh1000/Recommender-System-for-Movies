```python
#extracting genres from tmdb site.
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        #genre_str = " " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        #return genre_str.join(genres)
        return genres
    else:
        np.NaN
```

```python
df['genres'] = df['Title'].map(lambda x: get_genre(str(x)))
```

```python
#extracting Overview from tmdb site.
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_overview(x):
    overview = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['overview']:
        overview_str = "" 
        for i in range(0,len(data_json['overview'])):
            overview.append(data_json['overview'][i])
        return overview_str.join(overview)
        #return overview
    else:
        np.NaN
```

```python
df['overview'] = df['Title'].map(lambda x: get_overview(str(x)))
```

```python
#get director from Cast and crew feature
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    elif " (director/screenpaly)" in x:
        return x.split(" (director/screenpaly)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
df['director_name'] = df['Cast and crew'].map(lambda x: get_director(x))
```

```python
df['director_name']=df['director_name'].apply(lambda x: x.replace(" ",""))
```
