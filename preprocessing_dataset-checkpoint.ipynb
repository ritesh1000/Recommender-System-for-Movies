```python
#function frim ast i.e. literal_eval() is used to convert string to list
import ast
```

```python
def converter_to_list(x):
    l=[]
    for i in ast.literal_eval(x):
         l.append(i['name'])
    return l
```

```python
#converting the string of list of "genres" column to List.
movies['genres']=movies['genres'].apply(converter_to_list)
```

```python
movies.head(7)
```

```python
#converting the string of list of "Keywords" column to List.
movies['keywords']=movies['keywords'].apply(converter_to_list)
```

```python
movies.head(7)
```

```python
#processing Cast column
movies['cast'][0]
```

```python
def converter_cast_to_list(x):
    l=[]
    counter=0
    for i in ast.literal_eval(x):
        if counter <4:
            l.append(i['name'])
            counter=counter + 1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(converter_cast_to_list)
```

```python
movies.head(7)
```

```python
#preprocessing feature "Crew"
#from here, just need director name from the coulmn
```

```python
def converter_crew_to_list(x):
    l=[]
    for i in ast.literal_eval(x):
        if i["job"]=="Director":
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(converter_crew_to_list)
```

```python
movies.head(7)
```

```python
#preprocessing "overview" column
movies['overview'][0]
```

```python
#overview is a string column, converting it to the List
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(7)
```

```python
#Need to make value of features as single entity. Hence remove the spaces and join the two words to refer as one word.
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(7)
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(7)
```

```python
movies['tags']=movies['genres']+movies['keywords']+movies['overview']+movies['cast']+movies['crew']
```

```python
movies.head(7)
```

```python
#above all the features like genres/keywords/cast/crew/overview all combined in the "tages". Hence individual columns can be deleted.
movies=movies[["movie_id","title","tags"]]
```
