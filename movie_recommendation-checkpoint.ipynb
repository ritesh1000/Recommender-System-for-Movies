```python
import ast
def try_join(l):
    x=ast.literal_eval(l)
    for i in x:
        try:
            return ' '.join(x)
        except TypeError:
            return np.nan 
```

```python
new_data['tags']=new_data['tags'].apply(try_join)
```

```python
#convertin entire string to lower case
new_data['tags']=new_data['tags'].apply(lambda x:x.lower())
```

```python
#vectorization of text tags
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=7000,stop_words="english")
```

```python
vector=cv.fit_transform(new_data['tags']).toarray()
```

```python
#here in the dataset, words should be in it's root form. means steming needs to be done.
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem(new_data['tags'][0])
```

```python
new_data['tags']=new_data['tags'].apply(stem)
```

```python
#calculating cosine means calculating distance of each movie with every other movie. and creating matrix for the same.
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:8]
```

```python
#recommending movie
def recommend(movie):
    movie_index=new_data[new_data['title'] == movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:8]
    
    for i in movie_list:
          print(new_data.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
